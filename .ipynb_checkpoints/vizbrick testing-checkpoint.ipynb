```python
from rdflib import Graph
from flashtext import KeywordProcessor
import re

# Loadinb Brick Schema
g = Graph()
g.parse("/Users/slz/git_repo/vizbrick/Brick.ttl", format="turtle")
```

```python
def clean_keyword(string):
    new_string = re.sub(r"[^a-zA-Z0-9]","",keyword).lower()
    return new_string
```

```python
num=0

reverse_index = {}
for triple in g:
    if "definition" in triple[1]:
        class_name = triple[0].replace("https://brickschema.org/schema/Brick#","")
        definition = triple[2]
        parsed_definition = definition.split(" ")
        for keyword in parsed_definition:
            
            new_string = clean_keyword(keyword)
            try:
                reverse_index[new_string].append(class_name)
            except:
                reverse_index[new_string]=[class_name]
        if num==10:
            break
        num+=1
        
```
