```python
import pandas as pd
import numpy
import re
```

```python
df = pd.read_csv('NIST_meta.csv')
```

```python
df.head()
```

```python
def splitUppers(given_str):
    parsed = re.findall('[A-Z][^A-Z]*', given_str)
    return parsed
```

```python
def get_location_terms(df, col_name):
    vals = df[col_name].dropna().unique()
    vals_ = []
    for val in vals:
        val = val.replace(" "," ")
        if val=="MBedroom":
            val = "Master Bedroom"
        if val=="MBath":
            val = "Master Bathroom"
        if val=="Bath1":
            val = "Bathroom 1"
        if val=="Bath2":
            val = "Bathroom 2"
        if val=="Bedroom3":
            val = "Bedroom 3"
        if val=="Bedroom4":
            val = "Bedroom 4"
        if val=="Bedroom2":
            val = "Bedroom 2"
        vals_.append(val.strip())
    return list(set(vals_))

def get_data_label_terms(df, col_name):
    vals = df[col_name].dropna().unique()
    vals_ = []
    for val in vals:
        val = val.split("_")[1]
        vals_.append(val.strip())
    return list(set(vals_))

def get_terms(df, col_name):
    vals = df[col_name].dropna().unique()
    vals_ = []
    for val in vals:
        val = val.replace(" "," ")
        vals_.append(val.strip())
    return list(set(vals_))
```

```python
get_location_terms(df, 'Measurement Location')
```

```python
get_terms(df, 'Subsystem')
```

```python
list_of_data_label = get_data_label_terms(df, 'Data Label')
list_of_data_label
```

```python
dict_cnt = {}
for terms in list_of_data_label:
    parsed = splitUppers(terms)
    for term in parsed:
        try:
            dict_cnt[term]+=1
        except:
            dict_cnt[term]=1
print(dict_cnt)
```

```python
get_terms(df, 'Measured Parameter')
```

```python
def get_distinct_rel(col_a,col_b):
    vals = []
    for index, row in df.iterrows():
        a = str(row[col_a]).strip()
        b = str(row[col_b]).strip()
        if a!="nan" and b!="nan":
            vals.append(a+","+b)

    vals = list(set(vals))
    vals.sort()

    for rel in vals:
        parsed_rel = rel.split(",")
        a = parsed_rel[0]
        if a == "Bedroom 2":
            a = "Bedroom2"
        if a == "MBedroom":
            a = "Master Bedroom"
        b = parsed_rel[1]
        print(a,b)
```

```python
get_distinct_rel('Measurement Location','Subsystem')
```
